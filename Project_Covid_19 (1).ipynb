Q.3: Analyze the peak number of daily new cases in germany, france and italy. which country experience highest single day surge and when did it occur?

```python
countries1 = ['Germany', 'France', 'Italy']
## creating a dictionary key: country_name and value: (date, peak num of cases)
daily_new_cases = {}
for country in countries1:
  country_data = confirmed_cases[confirmed_cases['Country/Region'] == country]
  summed_daily_new_cases = country_data.iloc[:,4:].sum(axis = 0).diff().fillna(0)
  daily_new_cases[country] = summed_daily_new_cases
```

```python
daily_new_cases
```

```python
peak_record = {}
for country , data in daily_new_cases.items():
  peak_cases = data.max()
  peak_date = data.idxmax()
  peak_record[country] = (peak_cases, peak_date)
print(peak_record)
```

```python
plt.figure(figsize=(12,6))
for country , data in daily_new_cases.items():
  plt.plot(data.index , data.values, label = country)
plt.legend()
plt.title("daily new cases")
plt.xlabel('date')
plt.ylabel('no. of cases')
plt.show()
```

Q.4 :
Compare the recovery rates(recoveries/confirmed cases) between canada and australia as of dec 31,2020.
which country showed better management of pandemic according to this metric?

```python
recovery_rates = {}
countries2 = ['Canada', 'Australia']
for country in countries2:
  country_confirmed_cases = confirmed_cases[confirmed_cases['Country/Region']== country].iloc[:,4:].sum(axis = 0)
  country_recovered_cases = recovered_cases[recovered_cases['Country/Region']== country].iloc[:,4:].sum(axis = 0)
  recovery_rate = country_recovered_cases['12/31/20']/ country_confirmed_cases['12/31/20']
  recovery_rates[country] = recovery_rate
print(recovery_rates)
```

```python
plt.bar(recovery_rates.keys(), recovery_rates.values(), color = ['Blue','green'])
plt.title("Recovery Rate Comparision on 2020-12-31")
plt.ylabel("Recovery Rate")
plt.xlabel('country')
plt.show()
```

Q.5: What is the distribution of death rates(death/confirmed cases) among provinces in canada? Identify the province with the highest and lowest death rate as of the latest data point?

```python
death_cases_canada = death_cases[death_cases['Country/Region']=='Canada'].iloc[:,-1].sum(axis = 0)
confirmed_cases_canada = confirmed_cases[confirmed_cases['Country/Region']== 'Canada'].iloc[:,-1].sum(axis = 0)
```

```python
provinces_canada = confirmed_cases[confirmed_cases['Country/Region'] == 'Canada']['Province/State'].tolist()
death_rates = {}
for province in provinces_canada:
  province_confirmed_case = confirmed_cases[confirmed_cases['Province/State']== province].iloc[:,4:].sum(axis = 0 )
  province_death_cases = death_cases[death_cases['Province/State']== province].iloc[:,4:].sum(axis = 0)
  death_rate = province_death_cases['5/29/21']/ province_confirmed_case['5/29/21']
  death_rates[province]= death_rate
print(death_rates)

## Province with highest and lowest death rate

lowest = min(death_rates, key = death_rates.get)
highest = max(death_rates, key = death_rates.get)
```

```python
print(lowest,'&',highest)
```

```python
plt.figure(figsize=(12,6))
plt.barh(death_rates.keys(), death_rates.values(), color = 'lightpink')
plt.title('covid 19 death rates by Province (canada)')
plt.ylabel('Province')
plt.xlabel('Death Rate(%)')
plt.show()
```

Q.13: Analyze the ratio of recoveries to confirmed cases for thr Unites States on a monthly basis from March 2020 to May 2021. Which month experienced highest recovery ratio, ehat could be the potential reasons?


```python
final_merged_data
```

```python
US_df = final_merged_data[(final_merged_data['Country/Region']=='US') & (final_merged_data['Date'] >= '2020-03') & (final_merged_data['Date']< '2021-05')]
US_df
```

```python
US_df['Year-Month'] = US_df['Date'].dt.to_period('M')
```

```python
US_df
```

```python
monthly_recoveries = US_df.groupby('Year-Month')['Recovered'].sum()
monthly_confirmed =  US_df.groupby('Year-Month')['confirmed'].sum()

recover_rate_US = monthly_recoveries/monthly_confirmed
recover_rate_US
```

```python
max_value = recover_rate_US.max()
max_month = recover_rate_US.idxmax()

plt.figure(figsize=(12,6))
plt.plot(recover_rate_US.index.astype(str), recover_rate_US.values, marker= 'o' , color = 'green')
## highligh the highest point
plt.scatter(str(max_month), max_value , color='red', s= 250 , label = 'Highest recovery rate')
plt.title('Monthly Recovery Rate in united Sates')
plt.xlabel('Year-Month')
plt.ylabel('Recovery Rate')
plt.xticks(rotation = 45)
plt.legend()
plt.gca().set_facecolor('lightblue')
plt.gcf().set_facecolor('pink')
plt.grid(True)
plt.show()
```

Q.14: Analyze the monthly sum of confirmed cases, deaths, and recoveries for countries to understand the progression of the pandemic.

Q. 15 : Visualize monthly global waves of infection and recovery phases.

```python
# Aggregate global monthly totals
global_monthly = Monthly_sum.groupby('Year-Month')[['confirmed', 'Deaths', 'Recovered']].sum().reset_index()

plt.figure(figsize=(12,6))
plt.plot(global_monthly['Year-Month'].astype(str), global_monthly['confirmed'], label='Confirmed', marker='o')
plt.plot(global_monthly['Year-Month'].astype(str), global_monthly['Deaths'], label='Deaths', marker='o')
plt.plot(global_monthly['Year-Month'].astype(str), global_monthly['Recovered'], label='Recovered', marker='o')

plt.title('Global Monthly COVID-19 Trend')
plt.xlabel('Month')
plt.ylabel('Number of Cases')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Q.16 :  Analyze the monthly sum of confirmed cases, deaths, and recoveries for US,Italy and Brazil to understand the progression of the pandemic.

```python
filtered_data = final_merged_data[final_merged_data['Country/Region'].isin(['US', 'Italy', 'Brazil'])]
filtered_data['Year-Month'] = filtered_data['Date'].dt.to_period('M')
monthly_sum_trend_filtered = pd.pivot_table(filtered_data, index=['Country/Region', 'Year-Month'], values=['confirmed', 'Deaths', 'Recovered'],
aggfunc='sum')
```

```python
monthly_sum_trend_filtered
```

```python
import matplotlib.pyplot as plt

# Reset index to get columns for plotting
monthly_sum_trend_filtered = monthly_sum_trend_filtered.reset_index()
metrics = ['confirmed', 'Deaths', 'Recovered']

plt.figure(figsize=(12, 6))
for metric in metrics:
    for country in ['US', 'Italy', 'Brazil']:
        country_data = monthly_sum_trend_filtered[monthly_sum_trend_filtered['Country/Region'] == country]
        plt.plot(country_data['Year-Month'].astype(str),
                 country_data[metric],
                 label=f"{country} - {metric}")

plt.title('Monthly Trend of COVID-19 Metrics (US, Italy, Brazil)')
plt.xlabel('Month')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
